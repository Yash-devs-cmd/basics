## Data Cleaning and Pre-Processing


Data is Cleaned Beforehand


```python
df_cleaned = df.copy()
```

```python
df_cleaned.head()
```

```python
# check null values , check duplicates , check data types
df_cleaned.shape 
```

```python
df_cleaned.drop_duplicates(inplace=True)
```

```python
df_cleaned.shape # 1 duplicate row removed 
```

```python
df_cleaned.isnull().sum() # no null valus
```

```python
df_cleaned.dtypes
```

```python
df_cleaned['sex'].value_counts() # only 2 categories and no typos
```

## Label Encoding for sex and smoker columns


female = 1 male = 0


```python
df_cleaned['sex'] = df_cleaned['sex'].map({"male":0,"female":1})
```

```python
df_cleaned.head()
```

```python
df_cleaned['smoker'].value_counts()# only 2 are present
```

```python
df_cleaned['smoker'] = df_cleaned['smoker'].map({"no":0,"yes":1})
```

```python
df_cleaned.head()
```

```python
df_cleaned.rename(columns = {
    'sex':'is_female',
    'smoker':'is_smoker'
}, inplace=True)
```

```python
df_cleaned.head()
```

## One-Hot Encoding For Region Column


```python
df_cleaned = pd.get_dummies(df_cleaned , columns = ['region'],drop_first=True)
```

```python
df_cleaned.head()
```

```python
df_cleaned = df_cleaned.astype(int)
```

```python
df_cleaned.head()
```

## Feature Engineering and Extraction


### Bmi can be split into categories underweight, normal weight, overweight, obese


```python

df_cleaned['bmi_category'] = pd.cut(
    df_cleaned['bmi'],
    bins=[0, 18.5, 24.9, 29.9, float('inf')],
    labels=['Underweight', 'Normal', 'Overweight', 'Obese']
)
```

```python
df_cleaned.head()
```

##### One hot encoding on bmi_category column


```python
df_cleaned = pd.get_dummies(df_cleaned , columns = ['bmi_category'],drop_first=True)
```

```python
df_cleaned = df_cleaned.astype(int)
```

```python
df_cleaned.head()
```

### Standard Scaling on age, bmi and children columns


```python
from sklearn.preprocessing import StandardScaler
```

```python
cols = ['age','bmi','children']
scaler = StandardScaler()
df_cleaned[cols] = scaler.fit_transform(df_cleaned[cols])
```

```python
df_cleaned.head()
```

## Model Training And Evaluation

```python
x = df_cleaned.drop("charges",axis=1)
y = df_cleaned["charges"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
 x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
n = x_test.shape[0] # rows
p = x_test.shape[1] # cols
adjusted_r2_score = (1-((1-r2)*(n-1))/(n-p-1)) 
print("RMSE:", rmse)
print("R^2 Score:", r2)
print("Adjusted R^2",adjusted_r2_score)
```
